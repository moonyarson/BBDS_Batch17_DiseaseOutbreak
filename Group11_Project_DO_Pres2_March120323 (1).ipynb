```python
# Frequency of words used
from collections import Counter

count_dict = Counter(news_words)

print(count_dict)
```

```python
# Words Counts/frequency 
words_counts = count_dict.most_common(10)
words_counts = dict(words_counts)
words_counts
```

```python
# Number of headlines
num_headlines = len(lines)
num_words = len(news_words)

sorted_dict = dict(sorted(count_dict.items(), key=lambda item: item[1], reverse=True))

#print(sorted_dict)
words_list = []
ph_list = []
pw_list = []

for key,value in sorted_dict.items(): 
    words_list.append(key)
    percent_headlines = value / num_headlines
    percent_words = value / num_words
    #print(key, value, percent_headlines, percent_words)
    ph_list.append(percent_headlines)
    pw_list.append(percent_words)
    print(f'{key:10} {value:4}   {percent_headlines:0.3f}   {percent_words:0.3f}')
```

```python
import matplotlib.pyplot as plt #absolute count

plt.bar(words_counts.keys(),
        words_counts.values())
plt.xlabel("words")
plt.ylabel("frequency")
plt.title("News Headlines Frequency")
plt.xticks(rotation=90)

plt.show()
```
